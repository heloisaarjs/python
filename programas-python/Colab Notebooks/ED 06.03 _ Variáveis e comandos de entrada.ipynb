**VARIAVEIS DE MEMÓRIA (RAM)**

```python
nome = "Heloísa" #str / %s
idade = 19 #int / %i
salario = 1200.50 #float / %f
temFilhos = False #bool

#exibindo com contatenação
print(nome, "tem", idade, "anos", "e ganha R$",salario)

#exibindo com estrutura de dados
print("%s tem %i anos e ganha R$ %.2f" %(nome,idade,salario))
```

**ENTRADA DE DADOS**
