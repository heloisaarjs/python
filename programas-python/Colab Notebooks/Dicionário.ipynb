```python
#imprimindo dicionário
dic={"cor1":"azul","cor2":"rosa","cor3":"amarelo","cor1":"vermelho"}
print(dic)
#imprimindo chaves
print(dic.keys())
#imprimindo valores
print(dic.values())
#imprimindo itens
print(dic.items())
#incluindo cor
dic["cor4"] = "cinza"
print(dic)
#alterando conteúdo
dic["cor1"]="laranja"
print(dic)
#imprimindo apenas um valor
valor=dic["cor1"]
print(valor)
valor=dic.get("cor2")
print(valor)

for cor in dic.keys():
  print(cor)

for cor in dic.values():
  print(cor)

#deletando
del dic["cor1"] #deleta pelo índice
dic.pop("cor4") #deleta pelo método
print(dic)
dic.clear() #exclui tudo
print(dic)
```

Imagine um armazém que precisa gerenciar seu estoque de produtos. Este é um problema comum que pode ser abordado utilizando estruturas de dados como dicionários ou listas encadeadas. Por exemplo, você pode usar um dicionário onde as chaves são os códigos dos produtos e os valores são as quantidades disponíveis em estoque. Isso permite que você adicione, remova e atualize produtos de forma eficiente, garantindo um gerenciamento eficaz do estoque.

```python
frutas = {"maça":10, "abacate":15, "melancia":5, "melão":5, "limao":20, "tomate":15, "maracuja":25}
frutaUsu=input("Fruta:")
if frutaUsu in frutas:
   print(frutas.get("Quantidade em estoque:",frutaUsu))
else:
  print("Fruta não encontrada")
```
