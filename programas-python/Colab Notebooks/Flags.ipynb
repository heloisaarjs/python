Faça um programa que calcule o mostre a média aritmética de N notas

```python
#flag while nota != 999
totalNotas = 0
numNotas = 0
nota = float(input("Digite a nota ou 999 para encerrar: "))
while nota != 999:
  numNotas+=1
  totalNotas+=nota
  nota = float(input("Digite a nota ou 999 para encerrar: "))

media = totalNotas/numNotas
print(f"A média do aluno: {media}")
```
