```python
# Definindo a lista de estudantes
estudantes = []

# Função para adicionar um estudante
def adicionar_estudante(nome, notas):
    estudante = {
        "nome": nome,
        "notas": notas
    }
    estudantes.append(estudante)

# Função para calcular a média das notas de um estudante
def calcular_media(notas):
    return sum(notas) / len(notas)

# Função para exibir as informações dos estudantes
def exibir_estudantes():
    for estudante in estudantes:
        nome = estudante["nome"]
        notas = estudante["notas"]
        media = calcular_media(notas)
        print(f"Estudante: {nome}")
        print(f"Notas: {notas}")
        print(f"Média: {media:.2f}")
        print()

# Adicionando alguns estudantes
adicionar_estudante("Alice", [7.5, 8.0, 9.0])
adicionar_estudante("Bob", [6.0, 7.5, 8.5])
adicionar_estudante("Carlos", [8.5, 9.0, 9.5])

# Exibindo as informações dos estudantes
exibir_estudantes()
```
